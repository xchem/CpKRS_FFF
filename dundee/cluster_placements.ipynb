```python
df = iter2_poses.get_df(inspiration_score=True, distance_score=True, energy_score=True, inspiration_ids=True)
```

```python
df = iter2_poses.get_df(inspiration_score=True, distance_score=True, energy_score=True, inspiration_ids=True)
df["inspiration_ids"] = df["inspiration_ids"].apply(lambda x: tuple(x))
grouped = df.groupby("inspiration_ids", group_keys=False).apply(lambda g: g.nsmallest(5, "energy_score"))
pset = animal.poses[[int(i) for i in grouped.index.values]]
pset.add_tag("CpKRS_FrAncestor_bulkdock_iter2_cluster_best5")
pset
```

```python
df = iter1_poses.get_df(inspiration_score=True, distance_score=True, energy_score=True, inspiration_ids=True)
df["inspiration_ids"] = df["inspiration_ids"].apply(lambda x: tuple(x))
grouped = df.groupby("inspiration_ids", group_keys=False).apply(lambda g: g.nsmallest(5, "energy_score"))
pset = animal.poses[[int(i) for i in grouped.index.values]]
pset.add_tag("CpKRS_FrAncestor_bulkdock_iter1_cluster_best5")
pset
```
